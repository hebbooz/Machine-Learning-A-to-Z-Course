# Hierarchical Clustering

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Using the dendrogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(X, method='ward')) # ward method minimises variance within each cluster

plt.title('Dendrogram')
plt.xlabel('Customers (Observation Points)')
plt.ylabel('Euclidean Distance (Ward Method)')

plt.show()
```

## Training the Hierarchical Clustering model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering

hirarchical_clustering = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward') # affinity is the distance between two points, 
                                                                                                     # linkage is the method to minimise variance within each cluster
y_hirarchical_clustering = hirarchical_clustering.fit_predict(X) # y_hirarchical_clustering = dependent variable
```

```python
print(y_hirarchical_clustering) # what cluster each observation point belongs to
```

## Visualising the clusters

```python
plt.scatter(X[y_hirarchical_clustering == 0, 0], X[y_hirarchical_clustering == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hirarchical_clustering == 1, 0], X[y_hirarchical_clustering == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_hirarchical_clustering == 2, 0], X[y_hirarchical_clustering == 2, 1], s = 100, c = 'purple', label = 'Cluster 3')
plt.scatter(X[y_hirarchical_clustering == 3, 0], X[y_hirarchical_clustering == 3, 1], s = 100, c = 'green', label = 'Cluster 4')
plt.scatter(X[y_hirarchical_clustering == 4, 0], X[y_hirarchical_clustering == 4, 1], s = 100, c = 'orange', label = 'Cluster 5')

plt.title('Clusters of customers')

plt.xlabel('Annual Income ($K)')
plt.ylabel('Spending Score (1-100)')

plt.legend()

plt.show()

# can change the number of clusters to see which number of clusters is the best representation of the data
```
