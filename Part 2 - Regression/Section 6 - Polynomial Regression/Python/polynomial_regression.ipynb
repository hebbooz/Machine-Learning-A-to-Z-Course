# Polynomial Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values # changed the range due to the first column being redundant
y = dataset.iloc[:, -1].values
```

## Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()
linear_regressor.fit(X, y) # training the linear regression model on the whole dataset
```

## Training the Polynomial Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_regressor = PolynomialFeatures(degree=4) # degree is the power of the polynomial. Start with 2 and increase until the model fits the data

X_polynomial = polynomial_regressor.fit_transform(X)

linear_regressor_2 = LinearRegression()
linear_regressor_2.fit(X_polynomial, y) # training the polynomial regression model on the whole dataset
```

## Visualising the Linear Regression results

```python
plt.scatter(X, y, color='orange')
plt.plot(X, linear_regressor.predict(X), color='navy')

plt.title("Expected Salary (Linear Regression)")
plt.xlabel("Position")
plt.ylabel("Salary (millions)")

plt.show()
```

## Visualising the Polynomial Regression results

```python
plt.scatter(X, y, color='orange')
plt.plot(X, linear_regressor_2.predict(X_polynomial), color='navy') # Need to use the polynomial features here

plt.title("Expected Salary (Polynomial Regression)")
plt.xlabel("Position")
plt.ylabel("Salary (millions)")

plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```python
X_grid = np.arange(min(X), max(X), 0.1) # this is to make the curve smoother
X_grid = X_grid.reshape((len(X_grid), 1)) # reshape the array to a matrix

plt.scatter(X, y, color='orange')
plt.plot(X_grid, linear_regressor_2.predict(polynomial_regressor.fit_transform(X_grid)), color='navy') # Need to use the polynomial features here

plt.title("Expected Salary (Polynomial Regression)")
plt.xlabel("Position")
plt.ylabel("Salary (millions)")

plt.show()
```

## Predicting a new result with Linear Regression

```python
linear_regressor.predict([[6.5]]) # predicting a new result with linear regression. 
#Have to input observations as a 2D array. 
#First pair of square brackets is for the rows, second pair is for the columns
```

## Predicting a new result with Polynomial Regression

```python
linear_regressor_2.predict(polynomial_regressor.fit_transform([[6.5]])) # Using polynomial_regressor fit_transform to get the correct format for the prediction
```
