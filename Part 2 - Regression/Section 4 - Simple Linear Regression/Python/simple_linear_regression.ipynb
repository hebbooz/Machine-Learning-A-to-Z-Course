## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Import Dataset

```python
dataset = pd.read_csv('Salary_Data.csv') 
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, -1].values
```

## Split Dataset into Training/Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state = 0)
```

## Train Simple Linear Regression Model on Training Set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

regressor.fit(X_train, Y_train)
```

## Predict Test Set Results

```python
y_predicted = regressor.predict(X_test)
```

## Visualise Training Set Results

```python
plt.scatter(X_train, Y_train, color='red') #visualising the training set results
plt.plot(X_train, regressor.predict(X_train), color='blue') # X_train gives the predicted values of the training set

plt.title('Salary vs Experience (Training Set)')

plt.xlabel('Years of Experience')
plt.ylabel('Salary ($)')
```

## Visialise Test Set Results

```python
plt.scatter(X_test, Y_test, color='red') 
plt.plot(X_train, regressor.predict(X_train), color='blue') # Don't need to change this line as regressor is already trained on the training set

plt.title('Salary vs Experience (Test Set)')

plt.xlabel('Years of Experience')
plt.ylabel('Salary ($)')
```

```python
print(regressor.predict([[6]])) # Predicting the salary of a person with 6 years of experience
```

## Find Equation of Linear Regression

```python
coefficient = regressor.coef_
intercept = regressor.intercept_

print("Salary ($) = " + str(round(coefficient[0],3)) + " * Years of Experience + " + str(round(intercept, 3)))
```
